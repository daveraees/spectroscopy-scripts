## more complicated plot, combining the chirp-corrected datasets

```python
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

trngs=[[0.5,0.59],
       #[1,1.2],
       [10,20],
       #[100,200],
       [1000,2000]]

f, subGraphs = pp.subplots(nrows=len(trngs),ncols=2,gridspec_kw={'width_ratios':[1.5,1.0]})
i=0
f.set_size_inches(6.2,6.0)

####################### VIS probe spectra #######################
wl_range=[440,710]
wl_ex=[535,575]

panelLetters=['a','b','c','d','e','f','g','h','i','j','k']
scaleFact = 1000
kFact = 1.0
xticks = range(450,749,50)

XminorLocator = MultipleLocator(10)
for tRange in trngs:
    if i < len(subGraphs):
        ax = subGraphs[i,0]
        visSpec_neatPHT=plotRepSpecTA(BP2_TA_vis_nir_AN['neat P3HT;vis']*scaleFact,
                              title=None, wl_range=wl_range, wl_ex=wl_ex, 
                              trngs=[tRange],ax=ax, grid=False,legend=False,
                              xticks=xticks)
        visSpec_AuPHT=plotRepSpecTA(BP2_TA_vis_nir_AN['AuNP@P3HT;vis']*scaleFact,
                              title=None, wl_range=wl_range, 
                              wl_ex=wl_ex,  trngs=[tRange],ax=ax, 
                              grid=False,legend=False,
                              xticks=xticks)
        visDiffSpec=visSpec_AuPHT - kFact * visSpec_neatPHT
        visDiffSpec.plot(ax=ax,grid=False)
        if i==0:
            #ax.legend(['neat P3HT','AuNP@P3HT','AuNP@PS'],loc='best',title=('%.1f - %.1f ps' % tuple(tRange))) 
            ax.legend([],loc='upper center',title=('(%s) %.1f - %.1f ps' % (panelLetters[i],tRange[0],tRange[1])),frameon=False) 
        else:
            ax.legend([],loc='upper center',title=('(%s) %.1f – %.1f ps' % (panelLetters[i],tRange[0],tRange[1])),frameon=False) 
        ax.set_ylabel('$\Delta A$ (mOD)');
        ax.xaxis.set_minor_locator(XminorLocator)
        if i == len(subGraphs)-1:
            ax.set_xlabel('Wavelength (nm)',visible=True)
            ax.set_xticklabels(xticks,visible=True)
        else:
            ax.set_xlabel('Wavelength (nm)',visible=False)
            ax.set_xticklabels(xticks,visible=False)
    else:    
        pass
    i +=1
####################### NIR probe spectra #######################
# NIR TA spectra
wl_range=[850,1300]
wl_ex=[]

#panelLetters=['f','g','h','i','j','k']
scaleFact = 1000
xticks = [900,1000,1100,1200]

XminorLocator = MultipleLocator(50)
j=0
for tRange in trngs:
    if j < len(subGraphs):
        ax = subGraphs[j,1]
        
        nirSpec_neatPHT=plotRepSpecTA(BP2_TA_vis_nir_AN['neat P3HT;NIR']*scaleFact,
                              title=None, wl_range=wl_range, wl_ex=wl_ex,  
                              trngs=[tRange],ax=ax, grid=False,legend=False,xticks=xticks)
        nirSpec_AuPHT=plotRepSpecTA(BP2_TA_vis_nir_AN['AuNP@P3HT;NIR']*scaleFact,
                              title=None, wl_range=wl_range, wl_ex=wl_ex,  
                              trngs=[tRange],ax=ax, grid=False,legend=False,xticks=xticks)
        # interpolate the dataset of neat P3HT into timestamps of the AuNP@P3HT dataset
        nirSpec_neatPHT = pd.DataFrame(data=np.interp(nirSpec_AuPHT.index, 
                                                      nirSpec_neatPHT.index.values,
                                                      nirSpec_neatPHT.values[:,0]),
                                       index=nirSpec_AuPHT.index,columns=nirSpec_neatPHT.columns)
        # calculate and plot the difference between AuNP and P3HT:
        nirDiffSpec=nirSpec_AuPHT - kFact * nirSpec_neatPHT
        nirDiffSpec.plot(ax=ax,grid=False)
        # adjust plot properties
        ax.yaxis.tick_right()
        ax.yaxis.set_label_position("right")
        ax.xaxis.set_minor_locator(XminorLocator)
        if j==0:
            #ax.legend(['neat P3HT','AuNP@P3HT','AuNP@PS'],loc='best',title=('%.1f - %.1f ps' % tuple(tRange))) 
            #ax.legend([],loc='upper right',title=('(%s) %.1f - %.1f ps' % (panelLetters[i],tRange[0],tRange[1])),frameon=False ) 
            pass
        else:
            #ax.legend([],loc='upper right',title=('(%s) %.1f - %.1f ps' % (panelLetters[i],tRange[0],tRange[1])),frameon=False) 
            pass
        ax.legend([],loc='upper right',title=('(%s)' % (panelLetters[i],)),frameon=False ) 
        ax.set_ylabel('$\Delta A$ (mOD)');
        if j == len(subGraphs)-1:
            ax.set_xlabel('Wavelength (nm)',visible=True)
            ax.set_xticklabels(xticks,visible=True)
        else:
            ax.set_xlabel('Wavelength (nm)',visible=False)
            ax.set_xticklabels(xticks,visible=False)
    else:    
        pass
    i +=1
    j +=1
f.subplots_adjust(hspace=0.07,wspace=0.05)  

# write the image file in publication quality
#f.savefig('BP2_TA_vis+nir_AN.png',dpi=300,bbox_inches='tight')
```
