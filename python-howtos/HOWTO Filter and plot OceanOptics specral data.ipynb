```python
# read a previously averaged and combined spectral dataset
Q150204_BP_P3HT_DB_an_Au = pd.read_csv('Q150204_BP_P3HT_DB_an_Au-DirectAbs-averages.csv',index_col=0)
```

```python
# just plot the dataset as it was
Q150204_BP_P3HT_DB_an_Au.plot()
```

```python
# make some smoothing and observe the result in comparison with unfitered data, to fine-tune the filters

# smoothing
Q150204_BP_P3HT_DB_an_Au_smooth = smOOth(Q150204_BP_P3HT_DB_an_Au,thrshld=2e-3,rm_span=6,ewma_span=6)

# plotting:
ax=pp.axes()
Q150204_BP_P3HT_DB_an_Au.plot(ax=ax,style=['--'],alpha=0.5)
for col in Q150204_BP_P3HT_DB_an_Au_smooth.columns: # remove the rows with missing data after filter
    Q150204_BP_P3HT_DB_an_Au_smooth.loc[:,col].dropna().plot(ax=ax)

zeroAbs = Q150204_BP_P3HT_DB_an_Au_smooth.loc[750:800,'0'].mean()
Q150204_BP_P3HT_DB_an_Au_smooth=Q150204_BP_P3HT_DB_an_Au_smooth-zeroAbs

Q150204_BP_P3HT_DB_an_Au_smooth.to_csv('Q150204_BP_P3HT_DB_an_Au-DirectAbs-averages-smoothed.csv')

# plot the filtered dataset alone:
Q150204_BP_P3HT_DB_an_Au_smooth.plot()

# save the figure
# pp.savefig('Q150204_BP_P3HT_DB_an_Au-DirectAbs-averages-smoothed.png')
```
